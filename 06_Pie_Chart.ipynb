# PieChart
Pie charts show the size of items (called wedge) in one data series, proportional to the sum of the items.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals=[1400,600,300,410,250]
exp_labels= ['Home Rent','Food', 'Internet Bill', 'Car', 'Other Utilities'] 
```

```python
# plt.axis("equal") # if the circle is not centered and is in 3d form then we use this property
plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels,radius=1.5) # radius is used for the pie chart size the default is 1 size
plt.show()
```

```python
# our chart is good but not perfect as it shows only the names but not the percentage so for that we use 'autopct' function
```

```python
plt.pie(exp_vals, labels=exp_labels,radius=1.5, autopct='%0.1f%%', shadow= True)
plt.show()
```

```python
# another argument is explode, when we define it, then it gonna explode that specific pie by that specific amount
```

```python
plt.pie(exp_vals, labels=exp_labels,radius=1, autopct='%0.1f%%', shadow= False , explode=[0,0.1,0,0,0.1])
plt.show()
```

```python
# the pie by default start from the 1st given name i-e home rent, but we can change it
```

```python
plt.pie(exp_vals, labels=exp_labels,radius=1, autopct='%0.1f%%', shadow= False , explode=[0,0.1,0,0,0.1],startangle=90)
plt.show()
```
