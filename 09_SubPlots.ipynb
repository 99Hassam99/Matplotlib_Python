```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
year=[2014,2015,2016,2017]
income=[4000,4500,5300,4600]
expenses=[2800,3000,2800,3400]
```

```python
plt.subplot(2,1,1)
plt.bar(year,income)
plt.ylabel('Income in $')
plt.title("Flow Chart bar")
```

```python
plt.subplot(2,1,2)
plt.bar(year,expenses,color='orange')
```

```python
plt.bar(year,income, label='Income')
plt.bar(year,expenses,color='orange',width=0.5,align='edge', label="Expense")

plt.legend()
```
