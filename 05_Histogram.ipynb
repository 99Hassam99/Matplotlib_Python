# Histogram
Histogram are bar charts where the yaxis contains the frequency of a variable.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# when working with histogram we need only a single dimension array and not the names of the patient
blood_sugar = [113,99,88,125,93,97,89,111,77,89,91,130,155,90]
plt.hist(blood_sugar)
```

```python
# by default it will plot 10 bins, in histogram there is a concept called bin or a bucket which is nothing but a range
```

```python
blood_sugar = [113,99,88,125,93,97,89,111,77,89,91,130,155,90]
plt.hist(blood_sugar, bins=[80,100,125,160], rwidth=0.90, color='green')
```

```python
# for step or blueprint of the chart we can add type to it
```

```python
blood_sugar = [113,99,88,125,93,97,89,111,77,89,91,130,155,90]
plt.hist(blood_sugar, bins=[80,100,125,160], rwidth=0.90, color='green', histtype='step')
```

```python
blood_sugar_men = [95, 110, 85, 155, 105, 98, 115, 172, 108, 99,122,139, 112, 107]
blood_sugar_women = [90, 100, 88, 125, 95, 92, 160, 105, 98, 122, 97, 104,128,135]

plt.xlabel('sugar range')
plt.ylabel('No. of patient')
plt.title('Blood sugar Analysis')

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,90,100,120,131,149,150,160],rwidth=0.80, color=['orange', 'green'], 
         label=['men','women'],)
plt.legend()
```

```python
# for horizontal histogram add orientation
```

```python
blood_sugar_men = [95, 110, 85, 155, 105, 98, 115, 172, 108, 99,122,139, 112, 107]
blood_sugar_women = [90, 100, 88, 125, 95, 92, 160, 105, 98, 122, 97, 104,128,135]

plt.xlabel('sugar range')
plt.ylabel('No. of patient')
plt.title('Blood sugar Analysis')

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,90,100,120,131,149,150,160],rwidth=0.80, color=['orange', 'green'], 
         label=['men','women'], orientation= 'horizontal')
plt.legend()
```
