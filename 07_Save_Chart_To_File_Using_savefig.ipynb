```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals=[1400,600,300,410,250]
exp_labels= ['Home Rent','Food', 'Internet Bill', 'Car', 'Other Utilities'] 
```

```python
plt.pie(exp_vals, labels=exp_labels,radius=1, autopct='%0.1f%%', shadow= False , explode=[0,0.1,0,0,0.1])
plt.show()
```

# To Save it use saveFig() function

```python
plt.pie(exp_vals, labels=exp_labels,radius=1, autopct='%0.1f%%', shadow= False , explode=[0,0.1,0,0,0.1])

# it has been saved
plt.savefig('PieChart.png')
```

```python
# if the chart is not fit in the picture use bbox argument
```

```python
plt.pie(exp_vals, labels=exp_labels,radius=1, autopct='%0.1f%%', shadow= False , explode=[0,0.1,0,0,0.1])

# it has been saved
plt.savefig('PieChart.png', bbox_inches='tight')
```

```python
# for extra border spacing use pad inches
```

```python
plt.pie(exp_vals, labels=exp_labels,radius=1, autopct='%0.1f%%', shadow= False , explode=[0,0.1,0,0,0.1])

# it has been saved
plt.savefig('PieChart.png', bbox_inches='tight', pad_inches=1)
```

```python
# we can also make it a transparent PNG File so by that we can use it on websites etc
```

```python
plt.pie(exp_vals, labels=exp_labels,radius=1, autopct='%0.1f%%', shadow= False , explode=[0,0.1,0,0,0.1])

# it has been saved
plt.savefig('PieChart.pdf', bbox_inches='tight', pad_inches=1, transparent= False)
```
