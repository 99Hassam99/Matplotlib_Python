```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[0,1,2,3,4,5,6,7]
y=[32,35,42,48,46,41,43,37]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,"g+") # it will plot + instead of a line
```

```python
# we can join these ++ by simply typing -- with g+
plt.plot(x,y,"g+--")
```

```python
plt.plot(x,y,"g+-.")
```

```python
# we can also change change the + sign to diamond for example
plt.plot(x,y,"rD")
```

```python
plt.plot(x,y,"rD-.")
```

these were like short cuts (g++, rD-- etc), but we can define them also

```python
plt.plot(x,y, color='red',marker='D', linestyle='dashdot',linewidth=1)
```

```python
# we can also change the color by RGB values like x000000
```

```python
plt.plot(x,y, color='blue',marker='D', linestyle='dashdot',linewidth=1,markersize=10) 
# markersize will increase the size of Diamond shape
```

another property is ALPHA which conrtol transparency
It has 0-1 scale means like either we choose 1 or we choose 0
0 is for less transparency and 1 is for high transparency.

```python
plt.plot(x,y,color='green',alpha=0.5)
```

```python
plt.plot(x,y,color='green',alpha=1)
```

```python
plt.plot(x,y,color='green',alpha=0)
```

```python
plt.plot(x,y,color='green',alpha=0.1)
```
