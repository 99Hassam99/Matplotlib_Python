Using following static data for companies revenues,

company=['GOOGL','AMZN','MSFT','FB']

revenue=[90,136,89,27]

Plot a bar chart that looks like below

Save this chart as jpg file and then pdf file with 5 inch of padding


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue)
plt.show()
```

```python
plt.xticks(xpos,company)
plt.ylabel('Tech Revenue in Billion')
plt.title('US Tech Stocks')
plt.bar(xpos,revenue, label='Revenue')

plt.legend()
```

```python
plt.xticks(xpos,company)
plt.ylabel('Tech Revenue in Billion')
plt.title('US Tech Stocks')
plt.bar(xpos,revenue, label='Revenue')
plt.legend()
plt.savefig('bar.pdf',pad_inches=5)
```
