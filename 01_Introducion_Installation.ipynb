# Matplotlib
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK

#pip install matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline 
#magic in jupyter noteboob for rendering the charts fine
```

```python
x=[0,1,2,3,4,5,6,7]
y=[49.5,50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
```

```python
# we can also change the color of the line and width of the line, for example
plt.plot(x,y, color='red' , linewidth=4)
```

```python
# we can also change the linestyle
plt.plot(x,y, color='red' , linewidth=4, linestyle='dashdot')
```

```python
plt.plot(x,y, color='red' , linewidth=4, linestyle='dotted')
```

```python
# we can also mention x and y axis in the chart and also title
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, color='red' , linewidth=4, linestyle='dashdot')
```
