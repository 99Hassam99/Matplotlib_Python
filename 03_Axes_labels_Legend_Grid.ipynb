```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
day = [1,2,3,4,5,6,7]
max_t=[47,45,46,48,49,46,47]
min_t=[39,41,38,42,37,43,36]
avg_t=[43,42,41,45,43,44,41]
```

```python
plt.plot(day,max_t)
```

# this is like one data with one line but we have 3 different data's so for that

```python
plt.plot(day,max_t)
plt.plot(day,min_t)
plt.plot(day,avg_t)
```

```python
# this is now perfect but still it is not readable as we dont know what type of data does x and y axis represent
plt.xlabel('DAYS')
plt.ylabel('TEMPERATURE')
plt.title("Weather")
plt.plot(day,max_t)
plt.plot(day,min_t)
plt.plot(day,avg_t)
```

# It's good now but still it is not perfect as we do not know what these lines represent
For that we have to define a legend.

```python
plt.xlabel('DAYS')
plt.ylabel('TEMPERATURE')
plt.title("Weather")
plt.plot(day,max_t, label= "Max")
plt.plot(day,min_t, label= "Min")
plt.plot(day,avg_t, label= "Avg")

plt.legend()
```

```python
# the legend box is in lower left corner but we can change it
```

```python
plt.xlabel('DAYS')
plt.ylabel('TEMPERATURE')
plt.title("Weather")
plt.plot(day,max_t, label= "Max")
plt.plot(day,min_t, label= "Min")
plt.plot(day,avg_t, label= "Avg")

plt.legend(loc='upper right')
```
